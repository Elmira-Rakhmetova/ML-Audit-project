```
import pandas as pd
import numpy as np
```

**DATASETS IMPORT**

```
audit_risk = pd.read_csv('https://raw.githubusercontent.com/Elmira-Rakhmetova/ML-Audit-project/main/audit_risk.csv')
trial = pd.read_csv('https://raw.githubusercontent.com/Elmira-Rakhmetova/ML-Audit-project/main/trial.csv')
```

**DATA PREPARATION**

```
audit_risk.describe()
```

```
trial.describe()
```

In order to proceed with the analysis, we decided to merge the 2 datasets. To do this, we edited *trial* dataset (column names, values for Score_A and Score_B) so that it is consistent with *audit_risk* dataset.

```
trial.columns = ['Sector_score','LOCATION_ID', 'PARA_A', 'Score_A', 'PARA_B',
       'Score_B',  'TOTAL', 'numbers', 'Marks',
       'Money_Value', 'MONEY_Marks', 'District',
       'Loss', 'LOSS_SCORE', 'History', 'History_score', 'Score', 'Risk_trial' ]
```

```
trial['Score_A'] = trial['Score_A']/10
trial['Score_B'] = trial['Score_B']/10
```

```
same_columns = np.intersect1d(audit_risk.columns, trial.columns)
same_columns
```

```
merged_df = pd.merge(audit_risk, trial, how='outer', on = ['History', 'LOCATION_ID', 'Money_Value', 'PARA_A', 'PARA_B',
       'Score', 'Score_A', 'Score_B', 'Sector_score', 'TOTAL', 'numbers'])
merged_df.columns
```
